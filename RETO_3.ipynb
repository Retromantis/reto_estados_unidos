```python
# Procedemos a eliminar columnas no útiles

df_cleaned = df.drop(columns=['RowNumber', 'CustomerId', 'Surname'])
df_cleaned.head()
```

----> TRABAJAREMOS CON AGE Y ESTIMATED SALARY, POR LO TANTO IMPUTAMOS:
SUSTITUIMOS LOS "MISSINGS" POR VALORES VALIDOS 

```python
#SUSTITUIMOS LOS "MISSINGS" POR VALORES VALIDOS
#Utilizaremos la función KNNImputer de sklearn.impute

from sklearn.impute import KNNImputer

imputer = KNNImputer()
null_Colums =["Age","EstimatedSalary"]
df_cleaned[null_Colums]=imputer.fit_transform(df_cleaned[null_Colums])

df_cleaned.isnull().sum()
```

```python
#Eliminando filas duplicadas
df = df_cleaned.drop_duplicates()
df

# ----> EL DF SE MANTUVO IGUAL, POR LO TANTO, NO HABIAN FILAS DUPLICADAS
```

VER SI ES NECESARIO EL CODIGO DE ABAJO y DESCUBRIR EL PORQUE SE APLICA:

```python
#VERIFICAR, PORQUE LOS DATOS YA ERAN ENTEROS, ¿ERA NECESARIO?
# Convertimos las columnas al tipo de datos adecuado
df['HasCrCard'] = df['HasCrCard'].astype('int')
df['IsActiveMember'] = df['IsActiveMember'].astype('int')
df['Exited'] = df['Exited'].astype('int')
```

```python
# Información general del DataFrame después de la limpieza
df.info()
```

COMPARAR CON EL 'df.info()' DEL INCIO

```python
#CONVERSION DE VARIABLES CATEGORICAS A NUMERICAS (GENDER Y GEOGRAPHY)

columnas_categoricas = ['Gender', 'Geography']

# Aplica one-hot encoding
df_encoded = pd.get_dummies(df, columns=columnas_categoricas)

# El nuevo dataframe 'df_encoded' ahora contiene las columnas codificadas
df_encoded.head()
```

```python
#Separamos los datos en conjuntos de entrenamiento y prueba 
#Lo aplicamos al nuevo df (df_encoded)
from sklearn.model_selection import train_test_split

# Definir variables predictoras (X) y objetivo (y)
X = df_encoded.drop(columns=['Exited'])
y = df_encoded['Exited']  # --> Variable Objetivo

# Dividir los datos en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

<Estamos enfrentando un problema de *clasificación binaria* (predecir si un cliente abandona o no el banco), por lo que usaremos modelos de clasificación como:

-Regresión Logística
-Árboles de Decisión
-Random Forest
-Gradient Boosting (XGBoost)

```python
#APLICAMOS LOS DIFERENTES ALGORITMOS SELECCIONADOS PARA COMPARAR CUAL ES EL QUE MAS NOS CONVIENE:

from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Inicializar los modelos
log_reg = LogisticRegression()
tree_clf = DecisionTreeClassifier()
rf_clf = RandomForestClassifier()
gb_clf = GradientBoostingClassifier()

# Entrenar los modelos
log_reg.fit(X_train, y_train)
tree_clf.fit(X_train, y_train)
rf_clf.fit(X_train, y_train)
gb_clf.fit(X_train, y_train)

# Predecir en el conjunto de prueba
y_pred_log_reg = log_reg.predict(X_test)
y_pred_tree_clf = tree_clf.predict(X_test)
y_pred_rf_clf = rf_clf.predict(X_test)
y_pred_gb_clf = gb_clf.predict(X_test)

#Usamos métricas de evaluación para comparar los modelos:
def evaluate_model(y_test, y_pred):
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    return accuracy, precision, recall, f1

#Evaluamos cada modelo ---------> repasarlo para entenderlo mejor:
metrics_log_reg = evaluate_model(y_test, y_pred_log_reg)
metrics_tree_clf = evaluate_model(y_test, y_pred_tree_clf)
metrics_rf_clf = evaluate_model(y_test, y_pred_rf_clf)
metrics_gb_clf = evaluate_model(y_test, y_pred_gb_clf)

# Resultados
print("Logistic Regression: ", metrics_log_reg)
print("Decision Tree: ", metrics_tree_clf)
print("Random Forest: ", metrics_rf_clf)
print("Gradient Boosting: ", metrics_gb_clf)
```

OBSERVACIONES:

-La regresión logística tiene una baja sensibilidad (recall), lo que significa que no identifica bien los casos positivos.
-El árbol de decisión tiene un buen recall, pero su precisión es menor.
-Random Forest y Gradient Boosting tienen un equilibrio entre precisión y recall.

```python
#PROBANDOLO NUEVAMENTE, PERO AGREGANDO LOS HIPERPARAMETROS:

from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Inicializar los modelos
log_reg = LogisticRegression(max_iter=1000)
tree_clf = DecisionTreeClassifier()
rf_clf = RandomForestClassifier(n_estimators=200, max_depth=10)
gb_clf = GradientBoostingClassifier(learning_rate=0.01, n_estimators=200, max_depth=5)

# Entrenar los modelos
log_reg.fit(X_train, y_train)
tree_clf.fit(X_train, y_train)
rf_clf.fit(X_train, y_train)
gb_clf.fit(X_train, y_train)

# Predecir en el conjunto de prueba
y_pred_log_reg = log_reg.predict(X_test)
y_pred_tree_clf = tree_clf.predict(X_test)
y_pred_rf_clf = rf_clf.predict(X_test)
y_pred_gb_clf = gb_clf.predict(X_test)

#Usamos métricas de evaluación para comparar los modelos:
def evaluate_model(y_test, y_pred):
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    return accuracy, precision, recall, f1

#Evaluamos cada modelo ---------> repasarlo para entenderlo mejor:
metrics_log_reg = evaluate_model(y_test, y_pred_log_reg)
metrics_tree_clf = evaluate_model(y_test, y_pred_tree_clf)
metrics_rf_clf = evaluate_model(y_test, y_pred_rf_clf)
metrics_gb_clf = evaluate_model(y_test, y_pred_gb_clf)

# Resultados
print("Logistic Regression: ", metrics_log_reg)
print("Decision Tree: ", metrics_tree_clf)
print("Random Forest: ", metrics_rf_clf)
print("Gradient Boosting: ", metrics_gb_clf)
```

```python
# EJERCICIO 4: Algoritmo de clusterización
#Aplicando K-means para encontrar grupos homogéneos de clientes y describir las características de cada grupo

from sklearn.cluster import KMeans
import numpy as np

# Preparar los datos para la clusterización
X_cluster = df_encoded.drop(columns=['Exited'])

# Aplicar K-means
kmeans = KMeans(n_clusters=4, random_state=42)
kmeans.fit(X_cluster)

# Asignar etiquetas de cluster
df_encoded['Cluster'] = kmeans.labels_

# Describir características de cada grupo
clusters = df_encoded.groupby('Cluster').mean()

clusters
```
